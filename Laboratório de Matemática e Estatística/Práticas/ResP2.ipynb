## Março
O mês de **março** (<u>AO 1945</U>: *Março*) é o terceiro mês do ano no ***calendário gregoriano*** e um dos sete meses com 31 dias.  
O nome `março` surgiu na **Roma Antiga**, quando era o primeiro mês do ano e chamava-se *Martius*, de <u>Marte</u>, o deus romano da guerra. Em **Roma**, onde o clima é *mediterrânico*, março é o primeiro mês da ***primavera***, um evento lógico para se iniciar um ano novo, bem como para que se comece a temporada das campanhas militares.

### Datas comemorativas
- Terça-feira de **Carnaval** (3 de fevereiro a 9 de março em ano não bissexto ou 4 de feveiro a 9 de março em ano bissexto)
- 8 de março - *Dia Internacional da Mulher*
- 9 de março - *Dia Internacional do DJ*
- 21 de março
  1. *Dia Internacional contra a Discriminação Racial*
  2. *Dia Internacional da Síndrome de Down*
  3. *Dia da Árvore (segundo a **ONU**)*
- 22 de março - *Dia Mundial da Água*
