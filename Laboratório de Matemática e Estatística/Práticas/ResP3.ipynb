### Exercício 3

Calcule $\ \displaystyle\sum_{k=1}^{20} \dfrac{1}{2^k}\ $   com 6 casas decimais. 

```python
import sympy as s
k=s.symbols('k')
resultado=s.Sum(1/2**k, (k, 1, 20)).doit().evalf()
print(round(resultado, 6))
```
