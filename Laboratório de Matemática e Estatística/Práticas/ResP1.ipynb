Suponhamos que no momento em que Aquiles inicia a perseguição à tartaruga, esta se encontra num ponto mais avançado. Apesar de Aquiles correr mais depressa do que a tartaruga e portanto se aproximar desta cada vez mais, Zenão afirmava que Aquiles percorreria a distância do seu ponto de partida $A$ ao ponto de partida $B$ da tartaruga, esta entretanto ter-se-ia movido para $C$; quando Aquiles alcançasse a posição $C$ já a tartaruga estaria mais à frente, na posição $D$; e assim sucessivamente.

Usando conceitos estabelecidos cerca de 2500 anos depois, eis a explicação pela qual Aquiles alcança de facto a tartaruga: apesar do número de intervalos de tempo ser infinito, o tempo necessário para Aquiles alcançar a tartaruga não é necessariamente infinito.


Suponhamos que à tartaruga é dado um avanço de 70 metros e que esta se desloca à velocidade de $0.03$ m/s, enquanto Aquiles se move à velocidade de $6$ m/s. Quanto tempo decorre até que Aquiles alcance a tartaruga?
