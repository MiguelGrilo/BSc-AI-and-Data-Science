#### Minimização de função de uma variável (minimize_scalar())

***Os parâmetros da função minimize_scalar():***

`minimize_scalar(fun, bounds=None, args=(), method='brent', tol=None, options=None)`

`fun` - Função que aceita um argumento e retorna o valor da função a ser minimizada nesse ponto.

`bounds` - Lista que contém os limites inferior e superior do intervalo de busca para a variável independente.

`method` - Tupla que contém argumentos adicionais para a função fun.

`tol` - Tolerância para a convergência. O algoritmo de minimização é considerado convergido quando a mudança na estimativa do mínimo é menor que tol.

`options` - Dicionário que contém opções adicionais para o algoritmo de minimização.

### Exercício 3

Use a função `minimize_scalar` convenientemente para resolver o problema 

$$ x(x-2)(x-5)^{2} + 12 \rightarrow \min $$

1. Pelo método `brent`

2. Pelo método `bounded` no intervalo [-4,0]

3. Faça o gráfico de função e confirme os resultados.

```python
#Alínea 1
#Define a função 'f':
def f(x):
    return x * (x - 2) * (x - 5)**2 + 12
#Descobre a solução a partir do método 'Brent':
resultado_brent = opt.minimize_scalar(f, method='brent')

print("Output:\n", resultado_brent)
print("\n\nFormato de resposta:\nA solução que minimiza a função é:", resultado_brent.x, "\nO valor mínimo da função é:", resultado_brent.fun)
```

```python
#Alínea 2
#Descobre a solução a partir do método 'Bounded' no intervalo [-4, 0]:
resultado_bounded = opt.minimize_scalar(f, bounds=(-4, 0), method='bounded')

print("Output:\n", resultado_bounded)
print("\n\nFormato de resposta:\nA solução que minimiza a função é:", resultado_bounded.x, "\nO valor mínimo da função é:", resultado_bounded.fun)
```

```python
#Alínea 3 (Res. 1)
import matplotlib.pyplot as plt
#Define o intervalo de x e y:
intervalo_x = np.linspace(-5, 6, 400)
intervalo_y = f(intervalo_x)
#Esboço do gráfico:
plt.figure(figsize=(10, 6))
plt.plot(intervalo_x, intervalo_y, label='f(x) = x(x-2)(x-5)^2 + 12')
plt.scatter([resultado_brent.x, resultado_bounded.x], [resultado_brent.fun, resultado_bounded.fun], color='red', zorder=5)
#plt.text(resultado_brent.x, resultado_brent.fun, f'Brent: ({resultado_brent.x:.2f}, {resultado_brent.fun:.2f})', fontsize=12, color='red', ha='right')
#plt.text(resultado_bounded.x, resultado_bounded.fun, f'Bounded: ({resultado_bounded.x:.2f}, {resultado_bounded.fun:.2f})', fontsize=12, color='red', ha='right')
plt.title('Minimização da função')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.grid(True)

plt.show()
```

```python
#Alínea 3 (Res. 2)
import matplotlib.pyplot as plt
#Define o intervalo de x e y:
intervalo_x = np.linspace(-5, 6, 400)
intervalo_y = f(intervalo_x)
#Esboço do gráfico:
plt.plot(intervalo_x, intervalo_y, label='f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')

plt.legend()
plt.grid(True)

plt.show()
```

### Exercício 5

Resolva o problema de optimização $(x_{1}-2)^{2}+(x_{2}-5)^{2}\rightarrow \min $

1. Pelo método dos Gradientes Conjugados (CG)

2. Pelo método de Nelder-Mead

```python
#Alínea 1
#Define a função 'f' (função objetivo):
def f(x):
    return ((x[0] - 2)**2 + (x[1] - 5)**2)
#Ponto inicial para a otimização
x0=np.array([0, 0])
#Descobre a solução usando o método dos Gradientes Conjugados (CG)
resultado_cg = opt.minimize(f, x0, method='CG', tol=1e-5)

print("Output:\n", resultado_cg)
print("\n\nFormato de resposta:\nA solução que minimiza a função é:", resultado_cg.x, 
      "\nO valor mínimo da função é:", resultado_cg.fun)
```

```python
#Alínea 2
#Ponto inicial para a otimização
x0=np.array([0, 0])
#Descobre a solução usando o método de Nelder-Mead
resultado_nm=opt.minimize(f, x0, method='nelder-mead', tol=1e-5)

print("Output:\n", resultado_nm)
print("\n\nFormato de resposta:\nA solução que minimiza a função é:", resultado_nm.x, 
      "\nO valor mínimo da função é:", resultado_nm.fun)
```

##### Resposta: x_min = (x1=2, x2=5), fun(x_min)=0

`linprog(c, A_ub=None, b_ub=None, A_eq=None, b_eq=None,

`b_eq` - É um array 1-D que define os termos do lado direito das restrições de igualdade.

`bounds` -  É uma sequência de tuplas definindo os limites de cada variável de decisão. Cada tupla é do formato `(limite_inferior, limite_superior)`

`method` - É uma string que define o método a ser usado para resolver o problema. Os métodos suportados incluem 'interior-point', 'revised simplex', 'highs', 'highs-ds', 'highs-ipm', 'highs-runs', 'simplex' e 'highs-ipm-ds'. O método padrão é 'simplex'.

### Exercício 7

Resolva o problema de Programação Linear

$z=-x_{1}+4x_{2}\rightarrow \min $

$-3x_{1}+x_{2}\leq 6$

$-x_{1}-2x_{2}\geq 4$

$\space x_{2}\geq -3$

```python
#Define os coeficientes da função objetivo:
c=[-1, 4]       #Coefiecientes para -x1 + 4*x2
#Coeficientes das desigualdades:
A=[[-3, 1],     #Coeficientes para -3*x1 + x2 <= 6
   [1, 2]]      #Coeficientes para x1 + 2*x2 <= -4 (inicialmente -> -x1 - 2*x2 >= 4)
#Limites das desigualdades:
b=[6, -4]       #Limites correspondentes
#Limite da variável x1:
x1_bounds=(None, None)   # x1 não tem restrições
#Limite da variável x2:
x2_bounds=(-3, None)     # x2 >= -3
#Descobre a solução do problema de programação linear
resultado=opt.linprog(c, A_ub=A, b_ub=b, bounds=[x1_bounds, x2_bounds], method='highs')

if resultado.success:
    print("Output:\n", resultado)
    print("\n\nFormato de resposta:\nx1 =", resultado.x[0], 
          "\nx2 =", resultado.x[1],
         "\nValor mínimo de z=", resultado.fun)
else:
    print("Não foi possível encontrar uma solução:", resultado.message)
```

###### Resposta:
###### O ponto do mínimo x_min = {x1=2,  x2=-3} 
###### O valor da função no ponto do mínimo f(x_min) = -14

### Exercício 8

Resolva os problemas 

1. $f(x) = sin(x^{2}) - x \rightarrow \min \quad $ no intervalo $x \in [1,4]$

2. $f(x) = sin(x^{2}) - x \rightarrow \max \quad $ no intervalo $x \in [1,4]$

3. Esboce o gráfico de $f(x)$. Compare os extremos encontrados com o gráfico de f(x).


```python
#Alínea 1
#Define a função 'f':
def f(x): return np.sin(x**2) - x
#Define o intervalo de x:
intervalo=(1, 4)
#Descobre a solução do problema:
resultado_min=opt.minimize_scalar(f, bounds=intervalo, method='Bounded')

print("Output:\n", resultado_min)
print("\n\nFormato de resposta:",
      "\nx mínimo =", resultado_min.x, 
      "\nValor mínimo de f =", resultado_min.fun)
```

```python
#Alínea 2
#Define a função para 'f máximo'
def f_max(x): return (-1)* f(x)
#Define o intervalo de x:
intervalo=(1, 4)
#Descobre a solução do problema:
resultado_max=opt.minimize_scalar(f_max, bounds=intervalo, method='Bounded')

print("Output:\n", resultado_max)  #fun aparece como sendo positivo porém deve ser negativo
print("\n\nFormato de resposta:",
      "\nx máximo =", resultado_max.x, 
      "\nValor máximo de f =", -resultado_max.fun)
```

```python
#Alínea 3
import matplotlib.pyplot as plt
#Define os intervalos de x e y:
intervalo_x=np.linspace(1, 4, 400)
intervalo_y=f(intervalo_x)
#Esboço do gráfico:
plt.figure(figsize=(10, 6))
plt.plot(intervalo_x, intervalo_y, label='f(x) = sin(x^2) - x')
plt.scatter([resultado_min.x, resultado_max.x], [resultado_min.fun, -resultado_max.fun], color='red', zorder=5)
#plt.text(x_min, f_min, f'Min: ({x_min:.2f}, {f_min:.2f})', fontsize=12, color='red', ha='right')
#plt.text(x_max, f_max, f'Max: ({x_max:.2f}, {f_max:.2f})', fontsize=12, color='red', ha='right')
plt.title('Gráfico da função f(x) = sin(x^2) - x')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend()
plt.grid(True)

plt.show()
```

### Exercício 9  

Resolva o problema 

$f(x)=(x_{2}-x_{1}^2)^{2} + (1- x_{1})^{2} +10(x_{2}-1)^2 \rightarrow \min $

```python
#Define a função objetivo, 'f':
def f(x):
    return ((x[1] - x[0]**2)**2 + (1 - x[0])**2 + 10*(x[1] - 1)**2)
#Ponto inicial para a otimização:
x0=[0, 0]
#Descobre a solução para o problema de otimização:
resultado = opt.minimize(f, x0, method='Nelder-Mead')

print("Output:\n", resultado)
print("\n\nFormato de resposta:",
      "\nx1 =", resultado.x[0], 
      "\nx2 =", resultado.x[1], 
      "\nValor mínimo de f =", resultado.fun)
```

### Exercício 10

Resolva o problema

$72x_{1}+35x_{2}\rightarrow \min $

$5x_{1}+4x_{2}\geq 200$

$3x_{1}+x_{2}\geq 66$

$x_{1}+4x_{2}\geq 60$

$x_{1}\geq 0,$ $x_{2}\geq 0$

```python
#Coeficientes da função objetivo:
c=[72, 35]
#Coeficientes das desigualdades (Ax <= b)   
#É NECESSÁRIO INVERTER A EQUAÇÃO ORIGINAL EM TODOS OS CASOS
A=[[-5, -4], 
   [-3, -1], 
   [-1, -4]]
#Limites das desigualdades
b=[-200, -66, -60]
#Limites das variáveis x1 e x2:
x1_bounds=(0, None)   # x1 >= 0
x2_bounds=(0, None)   # x2 >= 0
#Descobre a solução para o problema de programação linear
resultado=opt.linprog(c, A_ub=A, b_ub=b, bounds=[x1_bounds, x2_bounds], method='highs')

if resultado.success:
    print("Output:\n", resultado)
    print("\n\nSolução encontrada:")
    print("x1 =", resultado.x[0])
    print("x2 =", resultado.x[1])
    print("Valor mínimo de 72x1 + 35x2 =", resultado.fun)
else:
    print("Não foi possível encontrar uma solução:", resultado.message)
```
