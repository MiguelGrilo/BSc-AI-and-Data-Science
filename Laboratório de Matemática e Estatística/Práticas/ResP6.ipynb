```python
import matplotlib.pyplot as plt
import numpy as np
def y1(x):
    return x * np.sin(x)
def y2(x):
    return np.sin(x)
intervalo_x = np.linspace(-6, 5, 1000)
intervalo_y1 = y1(intervalo_x)
intervalo_y2 = y2(intervalo_x)

plt.figure(figsize=(9,6))
plt.plot(intervalo_x, intervalo_y1, label='x*sin()', color='blue')
plt.plot(intervalo_x, intervalo_y2, label='sin()', color='orange', linestyle='--')

plt.title('Gráfico das 2 funções')
plt.xlabel('variável x')
plt.ylabel('variável y')
plt.legend()
plt.grid()
plt.xlim(-8, 8)
plt.ylim(-7, 7)
plt.axhline(color='grey')
plt.axvline(color='grey')

max1=-3*np.pi/2
plt.scatter(max1, np.sin(max1), color='green', marker='o', label='max1')
plt.annotate('max1', xytext=(-np.pi,4) ,arrowprops=dict(facecolor='gray'), xy=(-3*np.pi/2, 1))

max2=np.pi/2
plt.scatter(max2, np.sin(max2), color='blue', marker='o', label='max2')
plt.annotate('max2', xytext=(np.pi/2,4) ,arrowprops=dict(facecolor='gray'), xy=(np.pi/2, 1))

plt.legend()

plt.text(-7, 1, 'TAREFA 2')

plt.show()
```

```python
#Alínea 1
import numpy as np
def f(x):
    return x * np.sin(x) + np.log(x)
def primeira_derivada(x):
    return np.gradient(f(x), x)
def segunda_derivada(x):
    return np.gradient(np.gradient(f(x), x), x)
```

```python
# Color maps:
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=False)
a = np.outer(np.arange(0, 1, 0.01), np.ones(10))
plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=0.8, bottom=0.05, left=0.01, right=0.99)
maps = [m for m in plt.cm.datad if not m.endswith("_r")]  #figura com todos os mapas de cores disponíveis
                                                          #com sua instalação. O nome com '_r' anexado no 
                                                          #final lhe dará o mesmo mapa, com as cores na direção 
                                                          #oposta.
maps.sort()
l = len(maps) + 1
for i, m in enumerate(maps):
    plt.subplot(1, l, i+1)
    plt.axis("off")
    plt.imshow(a, aspect='auto', cmap=plt.get_cmap(m), origin="lower")
    plt.title(m, rotation=90, fontsize=10, va='bottom')
plt.show()
```

###  Exercício 9
**Campo dos gradientes**

<img src="https://services.meteored.com/img/tiles/cep010/5/15/12/037_viento@2x.png"  width="200px" height ='200px' >

(https://www.tempo.pt/mapas-do-tempo/viento-portugal.html)

Intuitivamente, podemos dizer que o gradiente é o vector que aponta para onde a grandeza resultante da função tem seu maior crescimento.

Formalizando,

O vector gradiente indica a taxa de variação de uma função escalar em relação à distância do deslocamento (deslocamentos infinitesimais).

$$\displaystyle \nabla =\frac {\text{distância vertical}}{\text{distância horizontal}}= \left(\frac {\partial }{\partial x_{1}},\frac {\partial }{\partial x_{2}},\cdots ,\frac {\partial }{\partial x_{n}}\right)$$


Consideremos $f:D\subset\mathbb{R}^n \mapsto \mathbb{R}$ um campo escalar diferenciável e um ponto $P=(x_1,...,x_n)\in D$.


O vector gradiente de $f$ no ponto $P$,$\nabla f(P)=\left(\frac {\partial f}{\partial x_{1}}(P),\frac {\partial f}{\partial x_{2}}(P),\cdots ,\frac {\partial f}{\partial x_{n}}(P)\right)$, dá a direção do aumento mais rápido de $f$ e é ortogonal à curva de nível $f(X)=k(=f(P))$ de $f$ em $P$. Portanto, se nos movermos na direção perpendicular, obtemos o aumento máximo.

Assim, a partir do campo escalar e do operador gradiente, constrói-se um campo vectorial, que associa a cada ponto do espaço o correspondente vector gradiente de $f$.



**Ortogonalidade com a curva de nível**
Seja o conjunto $\displaystyle C=\{(x,y)\in D: f(X)=k\}$ uma superfície de nível, onde $X=\left(x_1,...,x_n\right)$ uma função de um parâmetro $t$ tal que $\displaystyle X(0)=\left(x_1^0,...,x_n^0\right)$ 

Considerando 
$\displaystyle f(X(t))=f(x_1(t),...,x_n(t)\right)=k$ temos

$\displaystyle {\sum_{k=1}^n\frac {\partial f}{\partial x_i}}{\frac {dx_i}{dt}}=0$

Esta equação pode ser interpretada como o produto escalar do gradiente de $f$ por um vector tangente a $f$ em $\displaystyle (x_1(0),...,x_n(0))$ logo os dois vectores são perpendiculares entre si.

**Derivada direcional**
Como $f$ é uma função diferenciável, então existe a derivada direcional de $f$ na direção de qualquer vetor unitário $u$ de $\mathbb{R}^n$ e $$D_u f(X)=\nabla f(X) . u$$
(projeção escalar do vetor gradiente em u.)

O valor máximo da derivada direcional $D_u f(X)$  $|\nabla f(X)|$ e ocorre quando o vector $u$ tem a mesma direção do vetor gradiente $\nabla f(X)$.

Os vetores de gradiente são longos onde as curvas de nível estão próximas umas das outras e curtas onde as curvas estão mais afastadas. Isso ocorre porque o comprimento do vetor gradiente é o valor da derivada direcional de e as curvas de nível com espaçamento próximo indicam um gráfico íngreme.

Imagine uma bola de ping-pong que salta na mesa, no intervalo de tempo $[0,10]$, com uma trajectória definida pela função $sin(\pi t)$. Visualize a variação da bola, através de uma "animação" do deslocamento.

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

t = np.linspace(0, 10, 1000)
def trajetoria(t):
    return np.sin(np.pi * t)

fig, ax = plt.subplots()
ax.set_xlim(0, 10)
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel('Tempo')
ax.set_ylabel('Posição')
ax.set_title('Trajetória')

line, = ax.plot([], [])

def animate(i):
    x = t[:i]
    y = trajetoria(t)[:i]
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, animate, frames=len(t), interval=5, blit=True)

plt.show()
```

```python
#Alínea 1
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def f(x):
    return 1/(2*x*np.sin(x**2-x))

b=25*np.pi/41
delta=0.01

x1=b
x2=b+delta
y1=f(x1)
y2=f(x2)

m_sec=(y2-y1)/(x2-x1)
b_sec=y1-m_sec*x1

def df(x):
    return (-2*x*np.cos(x**2-x)-1/(2*x**2*np.sin(x**2-x)**2))

m_tan=df(b)
b_tan=f(b)-m_tan*b

intervalo_x=np.linspace(1.6, 2.2, 100)
valores_sec=m_sec*intervalo_x+b_sec
valores_tan=m_tan*intervalo_x+b_tan

fig, ax = plt.subplots()
ax.set_xlim(1.6, 2.2)
ax.set_ylim(0, 10)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Convergência das retas secante e tangente')

line_sec, = ax.plot([], [], color='blue', label='Reta Secante')
line_tan, = ax.plot([], [], color='red', label='Reta Tangente')
point, = ax.plot([], [], marker='o', color='black')

def animate(i):
    line_sec.set_data(intervalo_x[:i], valores_sec[:i])
    line_tan.set_data(intervalo_x[:i], valores_tan[:i])
    point.set_data(b, f(b))
    return line_sec, line_tan, point

ani = FuncAnimation(fig, animate, frames=len(intervalo_x), interval=50, blit=True)

plt.legend()
plt.grid()
plt.show()
```
