## Métodos Numéricos
# .......................................................


Os métodos iterativos geram uma sucessão convergente para a solução do problema: 

$x^{(0)}\rightarrow x^{(1)}\rightarrow x^{(2)}\rightarrow x^{(3)}\rightarrow
x^{(4)}\rightarrow ....\rightarrow x^{\ast }$

Os pontos  $x^{(k)}$ calculam-se por recorrência pela fórmula de iteração a partir do ponto inicial $x^{(0)}$ dado:

$x^{(k+1)}=F(x^{(k)})\quad$   -     *fórmula de iteração*

As iterações terminam quando se verifica o *critério de paragem*. Um critério de paragem típico é 

$\vert x^{(k+1)}-x^{(k)} \vert \leq \varepsilon \vert x^{(k)}\vert$ -  *critério de paragem*

onde $\varepsilon$ é uma *tolerância*, i.e. a constante que determina a precisão do resultado  $\vert x\vert$ - o valor absoluto de $x$.

Normalmente, a tolerância $\varepsilon=10^{-p}$ garante (p-1) casas decimais (significativas) certas na resposta. Por exemplo, com $\varepsilon=0.0001$ obtemos o resultado com 3 casas décimais (significativas) certas.

## Resolução de equações não lineares

## Método de Newton
O método de Newton aplica-se às equações não lineares na forma $\space f(x)=0$, para obter o valor aproximado da solução. Graficamente, a solução da equação é a abcissa do ponto de intersecção do gráfico de $f(x)$ com o eixo dos $xx$. O método consiste na aproximação da função $f(x)$ pela recta tangente ao gráfico de $f(x)$ no ponto $(x_{k},f\left(x_{k}\right)$ e encontrar a abcissa do ponto de intersecção da recta tangente com o eixo dos $xx$. Quando $x_{k}$ está próximo da solução exacta $x_{*}$, podemos esperar que os pontos de intersecção do gráfico de $f(x)$ e da tangente com o eixo dos xx estão próximos também.

Uma ilustração gráfica do funcionamento do método a partir do ponto $x_0$:

![](http://home.uevora.pt/~bushen/LabMat/A3-newton1.png)

No exemplo ilustrado, os pontos $x_1$, $x_2$, $x_3$,... gerados pelo método convergem para a solução exacta $x_{*}$.

Fórmula de iteração do método Newton:
$\displaystyle x_{k+1} = x_{k} - \frac{f(x_{k})}{f^{\prime}(x_{k})}$

Critério de paragem:

Para além do critério de paragem descrito anteriormente, podemos utilizar também o seguinte critério

$\vert x^{(k+1)}-x^{(k)} \vert \leq \varepsilon(\vert x^{(k)}\vert + \vert x^{(k+1)}\vert)$

### Aplicação do Método de Newton no calculo dos extremos (min, max) de uma função.

A condição necessária do extremo para a função f(x) é  $\space f^{\prime}(x)=0$.

Então, se $f$ é uma equação não linear, podemos aplicar o método de Newton para resolver a equação $\space f^{\prime}(x)=0$, chegando à fórmula de iteração 

$x_{k+1} = x_{k} - \frac{f^{\prime}(x_{k})}{f^{\prime\prime}(x_{k})}$.

**Observação**: 

>o método de Newton serve para calcular os mínimos e os máximos de $f(x)$ porque as condições necessárias em ambos os casos são identicas e pode ser classificado usando o valor da segunda derivada ($f''(x^{*})>0$ - mínimo, $f''(x^{*})<0$ - máximo). 

**Nota**
>O resultado $x^{*}$ depende do ponto inicial `x0`.

### Exercício 7  

1. Encontre os extremos locais de $f(x)=sin(x)+ln(x)-2$ pelo método de Newton com uma tolerância inferior a $\varepsilon =0.001$ iniciando as iterações a partir de:

$\quad\quad$ a) $x_0 = 7; \quad\quad $ b) $x_0 = 10$

Classifique os extremos obtidos (min ou max), usando a segunda derivada.

**Nota**
> Utilize as funções `df1()` e `df2()` para calcular as derivadas.

3. Utilize função **newton** do Python para confirmar os seus resultados.

````python
#Alínea 1
import numpy as np
import scipy.optimize as opt

def f(x):
    return np.sin(x) + np.log(x) - 2
def df1(x):
    return np.cos(x) + 1/x
def df2(x):
    return -np.sin(x) - 1/x**2
#Define a função que descobre os pontos críticos:
def encontrar_pontos_criticos(x0):
    pontos_criticos = opt.newton(df1, x0, fprime=df2, tol=0.001)
    return [pontos_criticos]
#Define a função que descobre os extremos locais:
def encontrar_extremos_locais(x0):
    pontos_criticos=encontrar_pontos_criticos(x0)
    extremos=[]
    for ponto in pontos_criticos:
        derivada2=df2(ponto)
        if derivada2>0:
            extremos.append((ponto, 'Minimo'))
        elif derivada2<0:
            extremos.append((ponto, 'Maximo'))
        else:
            extremos.append((ponto, 'Inconclusivo'))
    return extremos
````

## Método da bissecção

Pretende-se resolver a equação $f(x)=0$, supondo que $f(x)$ é **contínua** no intervalo $[a,b]$, e que se verifica a condição   

$$\quad f(a)f(b) < 0$$.

A existência da raíz de $f(x)=0$  é garantida, neste caso, pelo Teorema de Bolzano.

1. Implemente o método da bissecção para resolver a equação não linear $f(x)=0$ no intervalo [a,b]:

```
def bissec(f, a0, b0, niter):
    a = a0;  b = b0
    
    for k in range(niter):
        - calcular o ponto médio p do intervalo [a,b]
        - calcular os valores de função f(x) nos pontos a, p e b
        - se o valor absoluto de f(p) é inferior a constante eps=0.001, devolver p como a solução:
              return(p, eps) 
        - alterar [a,b] para [a,p], se  f(a)*f(p) < 0, e para [p,b] se  f(p)*f(b) < 0.
    
    calcular x = (a+b)/2
    calcular erro = (b-a)/2
    return (x, erro)       

```

2. Efetue 10 iterações do método da bissecção para resolver a equação não linear $\space sin(x)+ln(x)-2=0\space$ no intervalo [9,11].

2. No módulo **optimize** da biblioteca **scipy** existe a função **bisect**. Utilize os comando correctos para obter a informação sobre esta função.

3. Utilize função **bisect** do Python para confirmar os seus resultados.

````python
#Alínea 1
def bisseccao(f, a0, b0, niter):   #niter = número de iterações
    a=a0; b=b0
    
    for k in range(niter):
        p=(a + b) / 2
        fa=f(a)
        fp=f(p)
        fb=f(b)
        
        if abs(fp)<0.001:
            return p, abs(fp)
        elif fa*fp<0:
            b=p
        elif fp*fb<0:
            a=p      
    x=(a + b) / 2
    erro=(b - a) / 2
    return x, erro
````

````python
#Alínea 2
#Define a função 'f':
def f(x):
    return np.sin(x) + np.log(x) - 2
#Descobre a solução da bisseção de 'f' após 10 iterações no intervalo [9, 11]
resultado_bisseccao=bisseccao(f, 9, 11, 10)
print("Resultado da bissecção:", resultado_bisseccao)
````
