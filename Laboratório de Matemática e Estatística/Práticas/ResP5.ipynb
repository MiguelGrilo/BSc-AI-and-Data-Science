### Exerc√≠cio 2  

Podemos criar um array de v√°rios tipos, definindo o array com a op√ß√£o dtype que pode assumir os seguintes tipos:

`dtype = 'int32', 'int64', 'int', 'float' ,'complex','str','bool',... `

Crie um ciclo for para definir e visualizar o array 
$A=\left[\begin{array}{ccc}
1&2&3\\
5&6&7
\end{array}
\right]$ com os diferentes tipos. Para cada tipo diferente do array determine a dimens√£o e a mem√≥ria ocupada por cada elemento do array. Comente.

```python
#Al√≠nea 2
valores=[[1, 2, 3], [5, 6, 7]]
dtype=['int32', 'int64', 'float32', 'float64', 'complex64', 'complex128', 'str', 'bool']
for dtype in dtype:
    A=np.array(valores, dtype=dtype)
    print("\nTipo de Dados:", dtype)
    print("Array A:\n", A)
    #display(A)
    print("Dimens√£o de A:", A.ndim)
    print("Espa√ßo ocupado por cada elemento de A:", A.itemsize)
```

### Exerc√≠cio 6

Considere a fun√ß√£o $$f(x) = 1+e^{-0.2x}-x^{1.5}.$$

Para $i=1,...,9$, tais que $x_i \in X = [1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3.0]$ calcule
$df(X) = [f^{(5)}(x_i)]$.

```python
def f(x):
    return (1+np.exp(-0.2*x)-x**1.5)
X=np.arange(1.0, 3.1, 0.25)
ordem=5
df5=f(X)
for _ in range(ordem):
    df5 = np.gradient(df5, X)
print(df5)
```
