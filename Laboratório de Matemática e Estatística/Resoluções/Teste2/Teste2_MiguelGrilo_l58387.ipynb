Considere a função real de variável real $f$ dada, e o seu domínio. Sejam $g=f'$ e $h=f''$.

1. Calcule analiticamente, usando a biblioteca SymPy, a primeira e a segunda derivadas de $f$, isto é, as funções $g$ e $h$.
1. Defina as três funções $f$, $g$ e $h$ usando a biblioteca NumPy.
1. Usando a biblioteca Matplotlib, esboce numa mesma figura os gráficos das funções $f$, $g$ e $h$. A figura deve ter o título "f e as duas derivadas", legenda para os gráficos, eixos a cinzento, grelha, e rótulos dos eixos. As cores dos gráficos devem ser as indicadas em cada versão. A lista usada para os pontos das abessissas deve ter 150 entradas.

**Versão 2** 

$f(x)=\mathrm{sen}\bigl(\log(x)+2\bigr)-5\mathrm{cos}(x)$, $D_f=[1,e]$. Cores para os gráficos: $f$ a verde, $g$ a cor-de-laranja, e $h$ a azul.

```python
# COLOCAR A RESOLUÇÃO AQUI:
#Alínea 1:
import sympy as sp
#Define x
x=sp.symbols('x') 
#Define f:
f = sp.sin(sp.log(x) + 2) - 5 * sp.cos(x)
#Define a função g, primeira derivada:
g = sp.diff(f, x)
#Define a função h, segunda derivada:
h = sp.diff(g, x)

print("f(x) =", f)
print("g(x) =", g)
print("h(x) =", h)
```

```python
#Alínea 2:
import numpy as np
#Define a função f:
def f(x): return (np.sin(np.log(x) + 2) - 5 * np.cos(x))
#Define a primeira derivada com recurso a numpy:
def g(x): return np.gradient(f(x), x)
#Define a segunda derivada com recurso a numpy
def h(x): return np.gradient(g(x), x)
```

```python
#Alínea 3:
import matplotlib.pyplot as plt
#Define o intervalo de x:
intervalo_x=np.linspace(1, np.e, 150)
#Define os intervalos das funções:
intervalo_f=f(intervalo_x)
intervalo_g=g(intervalo_x)
intervalo_h=h(intervalo_x)
#Esboço do gráfico:
plt.figure(figsize=(10,6))
#f(x)
plt.plot(intervalo_x, intervalo_f, color='green', label='f(x)')
#g(x)
plt.plot(intervalo_x, intervalo_g, color='orange', label='g(x)')
#h(x)
plt.plot(intervalo_x, intervalo_h, color='blue', label='h(x)')
#Eixo
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.title('f e as duas derivadas')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

Considere a função real de variável real $q$ dada. 
1. Usando a função **newton** do módulo **optimize** da biblioteca **scipy**, aproxime um zero de $q$, a partir do ponto $x_0$ dado, e com a tolerância $\epsilon$ dada.
1. Implemente o método de bissecção no intervalo $[a,b]$ dado, com um máximo de $30$ iterações, e tolerância $\epsilon$. (Sem recorrer a funções específicas do Python, como **bisect**.)


**Versão 2**   
$q(x)=(x-1)(x+5)(x-7)$, $x_0=2$, $\epsilon=0,00001$, $[a,b]=[0,4]$.

```python
# COLOCAR A RESOLUÇÃO AQUI:
#Alínea 1
import scipy as sci
from scipy import optimize as opt
#Define q:
def q(x): return ((x-1)*(x+5)*(x-7))

#Define a primeira derivada de q:      ???
#def dq1(x): return np.gradient(q, x)  ???

#Define o ponto inicial
x0=2
#Define a tolerância:
tolerancia=0.00001
#Descobre o resultado:
resultado=opt.newton(q, x0, tol=tolerancia)
print("Solução para x0 = 2:", resultado)
```

```python
#Alínea 2
#Define o método de bissecação:
def bisseccao(f, a0, b0, niter):   #niter = número de iterações
    a=a0; b=b0
    tolerancia=0.00001
    for k in range(niter):
        p=(a + b) / 2
        fa=f(a)
        fp=f(p)
        fb=f(b)
        
        if abs(fp)<tolerancia:
            return p, abs(fp)
        elif fa*fp<0:
            b=p
        elif fp*fb<0:
            a=p      
    x=(a + b) / 2
    erro=(b - a) / 2
    return x, erro
#Descobre a solução da bisseção de 'f' até 30 iterações no intervalo [0, 4]
resultado_bisseccao=bisseccao(q, 0, 4, 30)

print("Resultado da bissecção:", resultado_bisseccao)
```

**Versão 1**   
$t(x_1,x_2) = (x_1-3)^{2} + (x_2+7)^{2}$ $\quad$ (método dos Gradientes Conjugados (CG)).


```python
# COLOCAR A RESOLUÇÃO AQUI:
from scipy import optimize as opt
#Define a função objetivo, 'f':
def t(x):
    return ((x[0] - 3)**2 + (x[1] + 7)**2)
#Ponto inicial para a otimização:
x0=[0, 0]
#Descobre a solução para o problema de otimização:
resultado = opt.minimize(t, x0, method='cg')

print("Output:\n", resultado)
print("\n\nFormato de resposta:",
      "\nx1 =", resultado.x[0], 
      "\nx2 =", resultado.x[1], 
      "\nValor mínimo de f =", resultado.fun)
```

Pretende-se analisar a ocorrência de defeitos de fabrico numa fábrica de bicicletas. Com os dados fornecidos, a partir de uma amostra de biclcletas fabricadas durante uma semana: 
1. Construa a tabela de frequências associada à informação (a tabela deve ter frequências absolutas, relativas, absolutas acumuladas, e relativas acumuladas);
1. Construa um gráfico de barras a partir das frequências absolutas.

**Versão 2**   

Verificou-se que 183 biclcletas não tinham defeitos, 31 tinham 1 defeito, 10 tinham 2 defeitos, 4 tinham 3 defeitos e 3 tinham 4 defeitos.

```python
# COLOCAR A RESOLUÇÃO AQUI:
#Alínea 1:
dados = {
    'Defeitos': [0, 1, 2, 3, 4],
    'Frequência Absoluta': [183, 31, 10, 4, 3]
}
#Frequência relativa
total = 231
dados['Frequência Relativa'] = [frequencia_absoluta / total for frequencia_absoluta in dados['Frequência Absoluta']]
#Frequência absoluta acumulada
freq_abs_acumulada = []
acumulado = 0
for frequencia_absoluta in dados['Frequência Absoluta']:
    acumulado += frequencia_absoluta
    freq_abs_acumulada.append(acumulado)
dados['Frequência Absoluta Acumulada'] = freq_abs_acumulada
#Frequência relativa acumulada
freq_rel_acumulada = []
acumulado = 0
for frequencia_relativa in dados['Frequência Relativa']:
    acumulado += frequencia_relativa
    freq_rel_acumulada.append(acumulado)
dados['Frequência Relativa Acumulada'] = freq_rel_acumulada

#Tabela de frequências
print("| Defeitos | F. Abs. |        F. Rel.       | F. Abs. Acumulada | F. Rel. Acumulada  |")
for i in range(len(dados['Defeitos'])):
    print("| {:8} | {:7} | {:20} | {:17} | {:^18} |".format(
        dados['Defeitos'][i],
       dados['Frequência Absoluta'][i],
       dados['Frequência Relativa'][i],
        dados['Frequência Absoluta Acumulada'][i],
        dados['Frequência Relativa Acumulada'][i]
    ))
```

```python
#Alínea 2:
#Gráfico de barras
plt.bar(dados['Defeitos'], dados['Frequência Absoluta'], color='blue')
plt.xlabel('Número de Defeitos')
plt.ylabel('Frequência Absoluta')
plt.title('Frequência de Defeitos em Bicicletas')

plt.show()
```
