**Versão 1** 

- Avaliar a expressão    $\displaystyle \dfrac{1}{x^{2}+7x+4}$  no ponto $\space x=4 \space$  pelas funções `.subs()` e `.evalf()`. 
    
- Calcular $\displaystyle \int_{0}^{\ln (5)}\sqrt{e^{x}-1}\,dx $     
    
- Calcular $\displaystyle \lim \limits_{x\rightarrow 1}\dfrac{\log (x)}{x-\sqrt{x}}$ 
    
- Encontrar o polinómio de Taylor de ordem 8 para $\displaystyle \space f(x)=x\,\cos^{2}(x)\space$ na vizinhança do ponto $\space a=\pi\space$       

```python
# COLOCAR A RESOLUÇÃO AQUI:
import sympy as sp
sp.init_printing()
x=sp.symbols('x')
def Expressao1(x):
    return (1/(x**2+7*x+4))
Expressao_no_ponto_4=Expressao1(x).subs(x, 4).evalf()
print("Ponto (1):", Expressao_no_ponto_4)

import math as m
def Expressao2(x):
    return (sp.sqrt(m.e**x - 1))
Integral_Expressao2_float = sp.integrate(Expressao2(x), (x, 0, sp.ln(5))).evalf()
print("Ponto (2):", Integral_Expressao2_float)

def Expressao3(x):
    return ((sp.log(x))/(x-sp.sqrt(x)))
limite_Expressao3 = sp.limit(Expressao3(x), x, 1)
print("Ponto (3):", limite_Expressao3)

def Expressao4(x):
    return (x*(sp.cos(x)**2))
a = sp.symbols('a')
a = sp.pi
def polinomio_Taylor(x):
    return (sp.series(Expressao4(x), x, a, 8))
print("Ponto (4):", polinomio_Taylor(x).subs(x, a))
```

$f(x,y) = -2x^{2} -4x -y^{4}+12y^{3}-54y^{2}+108y - 3$, $\qquad$ ponto  $p: (x,y)=(-1,3)$

**Versão 3**   

Pretende-se resolver o seguinte sistema linear pela inversão da matriz $A$, i.e. $x=A^{-1}b$, usando `scipy.linalg`

$$%
\begin{array}{c}
x_{1}-2x_{3}+x_{4}=2 \\ 
x_{2}+x_{3}+3x_4=2 \\ 
3x_{1}-x_{2}+x_{3}+4x_{4}=7 \\ 
2x_{1}+x_{2}+4x_{4}=9%
\end{array}%
$$

- Representar o sistema linear na forma matricial  $Ax=b$.
- Resolvê-lo pelo método indicado.

```python
# COLOCAR A RESOLUÇÃO AQUI:
import numpy as np
import scipy.linalg as sl
x1,x2,x3,x4=sp.symbols('x1 x2 x3 x4')
A=np.array([[1, 0, -2, 1], [0, 1, 1, 3], [3, -1, 1, 4], [2, 1, 0, 4]])
b=np.array([2, 2, 7, 9])
x=np.array([x1, x2, x3, x4])
print("A=",A,"\nx=",x,"\nb=",b)

inversa_A=sl.inv(A)
resultado=np.dot(inversa_A,b)
print("\nResultado:", resultado)
```
