### Conceitos fundamentais

###### Popula√ß√£o
Conjunto de elementos com uma caracter√≠stica comum, em estudo.

Uma popula√ß√£o pode ser <u>*finita*</u> ou <u>*infinita*</u>.

As medidas relativas a uma popula√ß√£o designam-se por <u>*par√¢metros*</u> que usualmente s√£o <u>*desconhecidos*</u> (mas fixos) e que se pretende conhecer.

######  Amostra
Subconjunto finito que se pretende representativo da popula√ß√£o.

As medidas relativas (**vari√°vel aleat√≥ria**) √† amostra designam-se por **estat√≠sticas** e variam de amostra para amostra.

### Estat√≠stica descritiva
formas de sintetizar a informa√ß√£o recolhida atrav√©s de tabelas, gr√°ficos e medidas estat√≠sticas, de modo a melhor caracterizar e interpretar essa informa√ß√£o.

* **Tabela de frequ√™ncias** - resume o conjunto de dados composto por $n$ observa√ß√µes. Fornece um acesso r√°pido ao n√∫mero, percentagem ou propor√ß√£o de elementos observados com uma determinada caracter√≠stica. Relaciona as vari√°veis nominais, vari√°veis ordinais, categorias ou classes com o n√∫mero de ocorr√™ncias (frequ√™ncias absolutas) e com a por√ß√£o (frequ√™ncia relativa) de observa√ß√µes de cada categoria, valor, intervalo de valores (classes de valores).

#### Dados qualitativos ou quantitativos discretos
A constru√ß√£o de uma tabela de frequ√™ncias para dados qualitativos ou quantitativos discretos depende da defini√ß√£o das seguintes colunas:

1¬™ Coluna: Todas as $ùêæ$ categorias ou valores distintos, $ùë•_ùëñ^\prime$, que os dados assumem;

2¬™ Coluna: As **frequ√™ncias absolutas**, $ùëõ_ùëñ, i=1,\ldots,K$, ou seja, o n√∫mero de vezes que cada categoria (valor) foi observada(o), sendo o n√∫mero total de observa√ß√µes $ùëõ=\sum\limits_{i=1}^K  ùëõ_ùëñ$ a dimens√£o do conjunto.

3¬™ Coluna: As **frequ√™ncias relativas**, $ùëì_ùëñ= \dfrac{ùëõ_ùëñ}{ùëõ}, i=1,\ldots,K$, ou seja, a propor√ß√£o de vezes que cada categoria (valor) foi observada(o);
Se os dados forem qualitativos ordinais ou quantitativos discretos s√£o acrescentadas as colunas:

4¬™ Coluna: As **frequ√™ncias absolutas acumuladas**, $ùëÅ_ùëñ= \sum\limits_{j=1}^i f_j$, ou seja, o n√∫mero de ocorr√™ncias das categorias (valores) inferiores ou iguais √† categoria (valor) atual;

5¬™ Coluna: As **frequ√™ncias relativas acumuladas**, $ùêπ_ùëñ=\dfrac{N_ùëñ}{ùëõ}= \sum\limits_{j=1}^i ùëõ_j$, ou seja, a propor√ß√£o de ocorr√™ncias das categorias (valores) inferiores ou iguais √† categoria (valor) actual.

**Nota**
> Para dados **qualitativos na escala nominal** <u>n√£o se apresentam</u> as frequ√™ncias (absolutas e relativas) acumuladas (4¬™ e 5¬™ colunas).

##### Tabela de frequ√™ncias para dados qualitativos ordinais ou quantitativos discretos.

||Categorias $ùë•_ùëñ^\prime$|frequ√™ncias absolutas $ùëõ_ùëñ$|frequ√™ncias relativas $f_ùëñ$ |frequ√™ncias absolutas acumuladas $N_ùëñ$|frequ√™ncias relativas acumuladas $F_ùëñ$ |

### Exerc√≠cio

Pretende-se analisar a ocorr√™ncia de acidentes de trabalho num determinado hospital. Em 397 profissionais de sa√∫de verificou-se que 16 n√£o sofreram qualquer acidente, 32 tiveram 1 acidente, 89
reportaram 2 acidentes, 137 sofreram 3 acidentes, 98 sofreram 4 acidentes e 25 profissionais reportaram 5 acidentes. 

Construa a tabela de frequ√™ncias associada √† informa√ß√£o 

||N√∫mero de Acidentes por profissional $ùë•_ùëñ^\prime$ |N√∫mero de Profissionais $ùëõ_ùëñ$|Propor√ß√£o de Profissionais $f_ùëñ$ |N√∫mero acumulado de Profissionais $N_ùëñ$|Propor√ß√£o acumulada de Profissionais $F_ùëñ$ |
| :----: | :----:  | :----:  | :-----: | :----:  | :----: |
| |   |    |   |   |   | 
| |    |    |   |    |  | 
| | | | | | | 
| |    |    |    |  |  | 
|Total   |  |      |        |    |  |

```python
Acidentes = [0,1,2,3,4,5]
Profissionais = [16,32,89,137,98,25]
n=len(Acidentes)
```

```python
def fi(seq):
    s=sum(seq)
    lfi = [i/s for i in seq]
    return lfi

def Ni(seq):
    s=sum(seq)
    lNi=[sum(seq[:i+1]) for i in range(len(seq))]
    return lNi

def Fi(seq):
    s=sum(seq)
    lFi=[sum(seq[:i+1])/s for i in range(len(seq))]
    return lFi
        
  
```

Esta tabela pode ser construida facilmente se fornecermos os dados como um DataFrame

```python
import pandas as pd
```

```python
dados=pd.DataFrame({'N√∫mero de Acidentes por profissional $ùë•_ùëñ^\prime$': Acidentes,
                    'N√∫mero de Profissionais $ùëõ_ùëñ$':Profissionais,
                    'Propor√ß√£o de Profissionais $f_ùëñ$':fi(Profissionais),
                    'N√∫mero acumulado de Profissionais $N_ùëñ$':Ni(Profissionais),
                    'Propor√ß√£o acumulada de Profissionais $F_ùëñ$':Fi(Profissionais)})
                
```

* ***Gr√°fico das frequ√™ncias acumuladas***

ou ***diagrama integral*** √© um gr√°fico de linhas onde s√£o
representadas as frequ√™ncias absolutas acumuladas $ùëÅ_ùëñ$ ou frequ√™ncias relativas acumuladas $ùêπ_ùëñ$. Este gr√°fico apresenta a
frequ√™ncia acumulada at√© cada uma das categorias/valores, notando que at√© √† primeira categoria/valor a frequ√™ncia acumulada √© nula. No gr√°fico das frequ√™ncias absolutas acumuladas, para categorias/valores superiores √†(ao) √∫ltima(o) a frequ√™ncia acumulada toma o valor $ùëõ$. No gr√°fico das frequ√™ncias relativas acumuladas, para categorias/valores superiores √†(ao) √∫ltima(o) a frequ√™ncia acumulada toma o valor 1. 

```python
# tratamento dos Dados

#import numpy as np

NNi=Ni(Profissionais).copy()
NNi.insert(0,0)  

FFi=Fi(Profissionais).copy()
FFi.insert(0,0)  

def fFA(x,cat,funclist):    
    condlist = [x < cat[0]]
    for i in range(len(cat)-1):
        condlist.append(x >= cat[i] and x<cat[i+1])
    condlist.append(x >= cat[-1])
    return np.piecewise(x,condlist,funclist)
    
           
x   = np.linspace(0,5.3,100)

y   = [fFA(i,Acidentes,NNi).tolist() for i in x]
yA = [NNi[i+1]-Profissionais[i] for i in range(n)]


y1= [fFA(i,Acidentes,FFi).tolist() for i in x]
y1A = [FFi[i+1]-fi(Profissionais)[i] for i in range(n)]



#Grafico
# import matplotlib.pyplot as plt


fig = plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.scatter(Acidentes, NNi[1:],marker='o',color="red")
plt.scatter(Acidentes,yA,color="none", edgecolor="red")
for i in Acidentes:
    plt.plot([i,i],[yA[i],NNi[i+1]], ls='--', c='blue' )
plt.plot(x, y, ls = '',marker='.')



plt.subplot(1, 2, 2)
plt.scatter(Acidentes, y1A,marker='o',color="none", edgecolor="red")
plt.scatter(Acidentes,FFi[1:],edgecolor="red")
for i in Acidentes:
    plt.plot([i,i],[y1A[i], FFi[i+1]], ls='--', c='blue' )

plt.plot(x, y1,ls = '',marker='.')

plt.show()   
```
