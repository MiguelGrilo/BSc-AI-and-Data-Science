# Jupyter notebooks

Jupyter notebooks consist of two main types of cells: code cells and text cells (Markdown).

1. **Code cells (default)**: These cells are where code is written and executed.
2. **Markdown cells**: These cells contain text formatted using Markdown (https://www.markdownguide.org/basic-syntax/)

The toolbar's dropdown menu allows changing the cell mode.



Alternatively, this can be done by entering the command mode with `Esc` and pressing `y` for code and `m` for text (Markdown). To edit a cell's content once again (and exit command mode), press `Enter`.

A list of available shortcuts can be found over `Help` > `Show Keyboard Shortcuts`.



## Running a cell

To execute code within a code cell (or show markdown in plain text), press the play button.



Shortcuts:

`Shift + Enter` runs the cell and moves on to the next one;

`Ctrl + Enter` runs the cell and stays on it;

`Alt + Enter` runs the cell and inserts a new one bellow.

Note: Code cells are executed in the order they are run. If a cell hasn't been executed earlier in the notebook, its results won't be available to the Julia interpreter.

## Adding a cell

To **add a cell** below the current one, click the `+` icon in the notebook's toolbar.



Alternatively, create a new cell below by pressing `Esc` and then `b`. To insert a cell above the current one, use `Esc` and then `a`.

## Deleting a cell

To **delete cells**, go to `Edit` > `Delete Cell`.



Another method would be to cut the cell with the scissor symbol.



The shortcuts for these are by pressing `Esc` followed by `d d` (twice to confirm), or `x` respectively.

Note: Jupyter lab allows for drag and drop cells, even among different notebooks (contrary to Jupyter notebook!)

## Saving a notebook

To save the notebook either press the save icon or go over at `File` > `Save Notebook`.





The `Ctrl+S` shortcut also works.

## Shutting down the kernel

To shut down the kernel hit the stop button and you can safely exit the notebook.



Note: restarting the kernel will delete all the variables from the environment and the cells must be rerun.
