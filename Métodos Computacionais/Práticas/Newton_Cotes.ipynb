```python
def trapezio(ff, a, b, alg=6):
    # Fórmula de Integração de Newton-Cotes de 2 pontos
    f = ff.function(x)
    I=((b-a)/2)*(f(a)+f(b))
    return N(I, digits=alg)
```

```python
f(x)=sqrt(x^2+4.);f
```

```python
trapezio(f, -1, 1)
```

```python
integrate(f, x, -1, 1).n()
```

```python
def plot_int_tra(func, a, b, step=0.1):
    # Ilustra graficamente a regra do trapézio
    x = var('x')
    f=func.function(x)
    P=line([(a,0), (a,f(a))], color='gray');
    P+=line([(a,f(a)), (b,f(b))], color='gray');
    P+=line([(b,0), (b,f(b))], color='gray');
    P+=polygon([(a,f(a)), (b,f(b))]+[(x, f(x)) for x in [a, a+step,...,b]]\
                +[(b,0),(a,0)], rgbcolor=(0.9,0.9,0.9), aspect_ratio='automatic');
    P+=plot(f,(a-0.5,b+0.5), thickness=1);
    return P
```

```python
trapezio(f,-1,1)
```

```python
plot_int_tra(f,-1,1)
```

```python
def simpson(ff, a, b, alg=6):
    # Fórmula de Integração de Newton-Cotes de 3 pontos
    f=ff.function(x)
    h=(b-a)/6
    pm=(a+b)/2
    I=h*(f(a)+4*f(pm)+f(b))
    return N(I, digits=alg)
```

```python
simpson(f, -1, 1)
```

```python
def tres_oitavas(ff, x0, x3, alg=6):
    # Fórmula de Integração das Três Oitavas
    f=ff.function(x)
    h=(x3-x0)/3
    x1=x0+h
    x2=x1+h
    I=((3*h)/8)*(f(x0)+3*f(x1)+3*f(x2)+f(x3))
    return N(I, digits=alg)
```

```python
f(x)=sqrt(x^2+4.);f
```

```python
tres_oitavas(f,-1,1)
```

```python
integrate(f(x),x,-1.,1.).n()
```

```python
def cinco_pontos(ff, x0, x4, alg=6):
    # Fórmula de Integração das Três Oitavas
    f=ff.function(x)
    h=(x4-x0)/4
    x1=x0+h
    x2=x1+h
    x3=x2+h
    x4=x3+h
    I=((2*h)/45)*(7*f(x0)+32*f(x0+h)+12*f(x0+2*h)+32*f(x0+3*h)+7*f(x0+4*h))
    return N(I, digits=alg)
```

```python
f(x)=sqrt(x^2+4.);f
```

```python
def NC_aberta_1pt(func, x0, xn, alg=6):
    # Fórmula de Integração de Newton-Cotes aberta do Ponto Médio
    f = func.function(x)
    a=x0
    b=xn
    I=(b-a)*f((a+b)/2)
    return N(I, digits=alg)
```

```python
f(x)=sqrt(x^2+4);x0=-1;xn=1
```

```python
NC_aberta_1pt(f,x0,xn)
```

```python
def NC_aberta_2pt(func, x0, xn, alg=6):
    # Fórmula de Integração de Newton-Cotes aberta com 2 pontos
    f = func.function(x)
    h=(xn-x0)/3
    I=((3*h)/2)*(f(x0+h)+f(x0+2*h))
    return N(I, digits=alg)
```

```python
NC_aberta_2pt(f,x0,xn)
```

```python
def NC_aberta_3pt(func, x0, xn, alg=6):
    # Fórmula de Integração de Newton-Cotes aberta com 3 pontos
    f = func.function(x)
    h=(xn-x0)/4
    I=((4*h)/3)*(2*f(x0+h)-f(x0+2*h)+2*f(x0+3*h))
    return N(I, digits=alg)
```

```python
NC_aberta_3pt(f,x0,xn)
```

```python
def NC_aberta_4pt(func, x0, xn, alg=6):
    # Fórmula de Integração de Newton-Cotes aberta com 4 pontos
    f = func.function(x)
    h=(xn-x0)/5
    I=((5*h)/24)*(11*f(x0+h)+f(x0+2*h)+f(x0+3*h)+11*f(x0+4*h))
    return N(I, digits=alg)
```

```python
NC_aberta_4pt(f,x0,xn)
```
