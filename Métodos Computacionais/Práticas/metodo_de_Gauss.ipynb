# Método da Gauss simples

```python
def eliminGauss(mat,vec):
    """
    Método da eliminação de Gauss
    simples
    
    Input:
    mat -> matriz dos coeficientes
    vec -> vector dos termos independentes
    
    Output:
    A -> matriz triangular superior
    b -> vector vec transformado
    """
    
    A = copy(mat)
    b = copy(vec)
    
    n = len(b)
    
    for k in [0..n-1]: # linha pivot
        
        for i in [k+1..n-1]: # percorre as linhas

            if abs(A[k,k]) < RR(1).ulp():
                return('Pivot nulo!')
            
            m = A[i,k]/A[k,k]
            
            for j in [k..n-1]: # percorre as colunas
                
                A[i,j]=A[i,j]-m*A[k,j]
                
            b[i] = b[i] - m*b[k]

    return(A,b)
```

```python
def subsAsc(A,b):
    """
    Resolucao de um sistema triangular
    superior pelo metodo das substituicoes
    ascendentes.
    
    Input:
    A -> matriz dos coeficientes
    b -> vector dos termos independentes
    
    Output:
    x -> solucao do sistema triangular
    """
    
    n = len(b)
    x = vector(RDF,n)
    
    for i in range(n-1,-1,-1):
        sum=b[i]
        for j in range(i,n):
            sum=sum-A[i,j]*x[j]
        x[i]=sum/A[i,i]
    return x
```

```python
def solveGauss(mat,vec):
    """
    Resolve um sistema de equacoes
    pelo metodo de Gauss simples

    Input:
    mat -> matriz dos coeficientes
    vec -> vector dos termos independentes
    
    Output:
    x -> solucao do sistema
    """
    
    T = copy(mat)
    v = copy(vec)
    
    T,v = eliminGauss(A,b)
    x = subsAsc(T,v)
    
    return x
```
