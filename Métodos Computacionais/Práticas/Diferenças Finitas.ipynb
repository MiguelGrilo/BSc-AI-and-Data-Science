```python
# Diferenças Finitas Progressivas
def df_progressivas(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(func(x0+h)-func(x0))/h
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Diferenças Finitas Regressivas
def df_regressivas(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(func(x0)-func(x0-h))/h
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Diferenças Finitas Centradas
def df_centradas(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(func(x0+h)-func(x0-h))/(2*h)
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Diferenças Finitas dos 3 Pontos (Progressiva)
def df_3p_progressiva(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(-3*func(x0)+4*func(x0+h)-func(x0+2*h))/(2*h)
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Diferenças Finitas dos 3 Pontos (Regressiva)
def df_3p_regressiva(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(3*func(x0)-4*func(x0-h)+func(x0-2*h))/(2*h)
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Diferenças Finitas da Regra dos 5 Pontos
def df_r5p(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df=(func(x0-2*h)-8*func(x0-h)+8*func(x0+h)-func(x0+2*h))/(12*h)
    df_exact=deriv(x0)
    error=abs(df-df_exact)
    error2=abs((df-df_exact)/df_exact)
    return N(df,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
# Aproximadamente Segunda Derivada
def df2_aproximada(f,x0,h,alg=6):
    func=f.function(x)
    deriv(x)=diff(diff(func(x),x))
    df2=(func(x0+h)-2*func(x0)+func(x0-h))/(h^2)
    df2_exact=deriv(x0)
    error=abs(df2-df2_exact)
    error2=abs((df2-df2_exact)/df2_exact)
    return N(df2,digits=alg), N(error,digits=alg), N(error2,digits=alg)
```

```python
def df_ppro_tol(f, x0, tol, alg=6):
    func=f.function(x)
    deriv(x)=diff(func(x),x)
    df_exact=deriv(x0)
    h=1.0
    er=1.
    while er > tol:
        h=0.25*h
        df=(-3.*func(x0)+4.*func(x0+h)-func(x0+2.*h))/(2.*h)
        er=abs(df-df_exact)
    return N(df_exact, digits=alg), N(df, digits=alg), N(er,digits=alg), N(h, digits=alg)
```
